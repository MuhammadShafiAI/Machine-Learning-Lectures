```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
    
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
```

```python
df
```

```python
df.columns = df.columns.str.strip()  # Remove leading/trailing spaces
df.columns = df.columns.str.replace(' ', '_')  # Replace spaces with underscores
df.columns = df.columns.str.replace(r'\(.*\)', '', regex=True)  # Remove parentheses and their contents
print(df.columns)
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per capita income')
plt.scatter(df['year'], df['per_capita_income_'], color="red", marker="+")
```

```python
reg= linear_model.LinearRegression()
reg.fit(df[['year']],df.per_capita_income_)
```

```python
reg.predict([[2024]])
```
